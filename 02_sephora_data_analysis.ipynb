## 세포라 온라인스토어(2023.3) 데이터 분석

### 📋제품별 데이터

#### 🔗제품 데이터 확인
> - product_id: 상품에 대한 고유 식별자 ID
> - product_name: 상품의 전체 이름
> - brand_id: 상품 브랜드에 대한 고유 식별자 ID
> - brand_name: 브랜드 전체 이름
> - loves_count: 이 제품을 즐겨찾기로 등록한 사람의 수
> - rating: 사용자 리뷰를 기반으로한 제품의 평균 평점(0.0~5.0)
> - reviews: 상품에 대한 사용자 리뷰 수
> - size: 상품 유형에 따라 oz, ml, g, 팩 또는 기타 단위로 표시될 수 있는 제품 크기
> - variation_type: 상품의 선택사항 매개변수 유형(예: 크기, 색상)
> - variation_value: 상품의 선택사항 매개변수의 구체적인 값(예: 100ml, Golden Sand)
> - variation_desc: 상품의 변형 매개변수에 대한 설명(예: 가장 밝은 피부톤)
> - ingredients: 재료, 제품에 포함된 성분 목록
> - price_usd: (미국 달러) 판매 가격으로 표시된 제품 가격
> - value_price_usd: 사이트의 가격 옆에 표시되는 상품의 정상 가격(가치가격)
> - limited_edition: 해당 상품이 한정판인지 여부 (1: True, 0:False)
> - new: 해당 상품이 신상인지 여부
> - online_only: 온라인에서만 판매되는지에 대한 여부
> - out_of_stock: 해당 상품의 품절 여부
> - sephora_exclusive: 세포라 독점 상품인지에 대한 여부
> - highlights: 상품의 속성을 강조하는 태그 또는 기능 설명
> - primary_category: 첫 번째 카테고리
> - secondary_category: 두 번째 카테고리
> - tertiary_category: 세 번째 카테고리
> - child_count: 사용 가능한 제품의 변형 수
> - child_max_price: 제품 변형 중 가장 높은 가격
> - child_min_price: 다양한 제품 중 최저가

```python
import pandas as pd

sephora_df = pd.read_csv("./datasets/product_info.csv")
display(sephora_df)
```

#### 🔗제품 데이터 전처리

```python
# 불필요한 컬럼 삭제하기
sephora_df = sephora_df.drop(columns=['product_name', 'size', 'variation_type', 'variation_value', 'variation_desc', 'ingredients', 'highlights', 'child_count', 'sale_price_usd',
                           'child_max_price', 'child_min_price'])
```

```python
sephora_df.info()
```

```python
# 중복행 검사
sephora_df.duplicated().sum()
```

```python
# 결측치 검사
sephora_df.isnull().sum()
```

```python
# 'reviews' 열에서 누락된 값을 '0'으로 대체
sephora_df['reviews'].fillna(0, inplace=True)
```

```python
# 'rating' 열에서 누락된 값을 '0'으로 대체
sephora_df['rating'].fillna(0, inplace=True)
```

```python
# value_price_usd 열의 빈 값을 같은 product_id의 price_usd 값으로 대체
sephora_df['value_price_usd'].fillna(sephora_df.groupby('product_id')['price_usd'].transform('mean'), inplace=True)
```

```python
# 두번째 카테고리가 빈값일 경우 첫번째 카테고리로 대체

def fill_secondary_category(row):
    if pd.isnull(row['secondary_category']):
        return sephora_df.loc[sephora_df['product_id'] == row['product_id'], 'primary_category'].iloc[0]
    else:
        return row['secondary_category']
sephora_df['secondary_category'] = sephora_df.apply(fill_secondary_category, axis=1)
```

```python
# 세번째 카테고리가 빈값일 경우 첫번째 카테고리로 대체

def fill_secondary_category(row):
    if pd.isnull(row['tertiary_category']):
        return sephora_df.loc[sephora_df['product_id'] == row['product_id'], 'primary_category'].iloc[0]
    else:
        return row['tertiary_category']
sephora_df['tertiary_category'] = sephora_df.apply(fill_secondary_category, axis=1)
```

```python
# 결측치 재확인
sephora_df.isnull().sum()
```

```python
# 할인율 계산해서 discount_rate 컬럼 추가하기
try:
    sephora_df['discount_rate'] = ((sephora_df['value_price_usd'] - sephora_df['price_usd']) / sephora_df['value_price_usd']) * 100
    # 음수인 경우 0으로 처리
    sephora_df['discount_rate'] = sephora_df['discount_rate'].apply(lambda x: max(x, 0))
    # inf인 경우 처리
    sephora_df['discount_rate'] = sephora_df['discount_rate'].replace(float('inf'), 0)
    # 소수점 둘째 자리까지 표시
    sephora_df['discount_rate'] = sephora_df['discount_rate'].round(1)
except Exception as e:
    print(e)
```

#### 📌관심도, 리뷰수, 평점의 상관관계 분석
> 관심도: 제품 구매 유무와 별개로 모든 고객이 표현할 수 있는 수치  
> 리뷰수: 구매 고객이 남길 수 있는 후기 개수  
> 평점: 리뷰를 남긴 고객이 매길 수 있는 등급으로 0부터 5까지 존재  
- 관심도와 리뷰 수는 서로 상관관계가 있지만, 평점은 연관이 가장 낮다.
- 관심도가 높은 상품일 수록 리뷰 개수가 많으며, 이는 관심도가 높은 제품일 수록 판매가 많이 되었음을 보여준다.
- 하지만 리뷰의 지표는 만족와 다르며, 리뷰 수가 많더라도 평점이 반드시 높지 않다는 것을 보여준다.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 카테고리별 고객 선호도, 리뷰수, 평점

# 각 카테고리별 고객 선호도, 리뷰수, 평점 계산
love_counts = sp_df.groupby('primary_category')['loves_count'].count().reset_index(name='Love_counts')
review_counts = sp_df.groupby('primary_category')['reviews'].count().reset_index(name='Review_counts')
grade_counts = sp_df.groupby('primary_category')['rating'].count().reset_index(name='Rating_counts')

order = ['Skincare', 'Makeup', 'Hair', 'Fragrance', 
        'Bath & Body', 'Mini Size', 'Men', 'Tools & Brushes', 'Gifts']

fig, axes = plt.subplots(3, 1, figsize=(10, 8))
plt.rcParams.update({'font.size': 7})

# 선호도 개수 시각화
sns.barplot(ax=axes[0], x='primary_category', y='Love_counts', data=love_counts, palette='Paired', order=order)  
axes[0].set_xlabel('Category')
axes[0].set_ylabel('Love Counts')

# 리뷰수 개수 시각화
sns.barplot(ax=axes[1], x='primary_category', y='Review_counts', data=review_counts, palette='Paired', order=order)
axes[1].set_xlabel('Category') 
axes[1].set_ylabel('Review Counts')

# 평점 개수 시각화
sns.barplot(ax=axes[2], x='primary_category', y='Rating_counts', data=grade_counts, palette='Paired', order=order)
axes[2].set_xlabel('Category')
axes[2].set_ylabel('Rating Counts')

# 그래프 출력
plt.tight_layout()
plt.show()
```

- 선호도, 리뷰수, 평점 모두 Skincare, Makeup, Hair, Fragrance 순서로 수치가 높은 것을 보여준다.
- 그 이하는 수치가 현저히 낮으므로 상위 카테고리 제품을 위주로 마케팅 전략을 짜는 것이 좋다.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 브랜드별 고객 선호도, 리뷰수, 평점

# 각 브랜드별 고객 선호도, 리뷰수, 평점 개수 계산 (상위 10개 브랜드)
love_counts = sp_df.groupby('brand_name')['loves_count'].count().reset_index(name='Love_counts')
love_counts = love_counts.sort_values(by='Love_counts', ascending=False).head(10)

review_counts = sp_df.groupby('brand_name')['reviews'].count().reset_index(name='Review_counts')
review_counts = review_counts.sort_values(by='Review_counts', ascending=False).head(10)

grade_counts = sp_df.groupby('brand_name')['rating'].count().reset_index(name='Rating_counts')
grade_counts = grade_counts.sort_values(by='Rating_counts', ascending=False).head(10)


fig, axes = plt.subplots(3, 1, figsize=(14, 10))
plt.rcParams.update({'font.size': 9})

# 선호도 개수 시각화
sns.barplot(ax=axes[0], x='brand_name', y='Love_counts', data=love_counts, palette='muted')  
axes[0].set_xlabel('Brand')
axes[0].set_ylabel('Love Counts')

# 리뷰수 개수 시각화
sns.barplot(ax=axes[1], x='brand_name', y='Review_counts', data=review_counts, palette='muted')
axes[1].set_xlabel('Brand') 
axes[1].set_ylabel('Review Counts')

# 평점 개수 시각화
sns.barplot(ax=axes[2], x='brand_name', y='Rating_counts', data=grade_counts, palette='muted')
axes[2].set_xlabel('Brand')
axes[2].set_ylabel('Rating Counts')

# 그래프 출력
plt.tight_layout()
plt.show()
```

#### 📌 마케팅 전략별 고객의 선호도, 리뷰수, 평점과의 상관관계
> 각 마케팅 전략은 한정판, 신상, 온라인 독점 판매, 세포라 독점 판매, 품절 여부를 나타낸다.  
> 고객의 선호도와 리뷰수는 마케팅 전략에 상관관계가 거의 없음을 보여준다.  
> 이는 화장품의 특성상 특정한 전략보다는 고객의 취향과 구매의사가 더 중요함을 알 수 있다.  

> 하지만 평점은 각 마케팅 전략에 따라 분포의 움직임이 다름을 보여주고 있다.  
> 평점이 아주 낮거나, 아주 높은 경우인데 이는 개인의 필요에 의해서가 아닌 마케팅 전략으로 인해 구매를 했을때,  
> 아주 만족하거나 아주 만족하지 못했음을 나타낸다.

> 결론으로는 억지로 마케팅 전략을 했을때보다 고객이 구매를 원하는 상품에 마케팅 전략을 더했을 때 매출 증진 효과를 기대할 수 있다.

  > 브랜드별 상위권 제품들이 상위권 카테고리에 속해있지만, 카테고리별 상위 제품을 분석했을때 상위 브랜드는 없었음.  
  > 브랜드마다 선호 제품과 비선호 제품이 뚜렷하기 때문에 카테고리별로 잘 나가는 제품들은 선별하여 제품 소진시 재구매를 유도할 수 있는 전략이 필요.

  > 타 브랜드 대비 세포라 자사 브랜드에 대한 고객 선호도와 관심도가 높음  
  > 단독 입점 및 한정판에 대한 마케팅 전략이 효과가 낮으므로 자사 제품을 강화하여 매출을 기대.  략

  > 신상품, 품절, 단독입점, 기획 등 다양한 마케팅 전략은 투자 대비 효과가 낮음.
  > 새로운 제품을 마케팅 할 때에는 이러한 프로모션 전략보다는 고객의 니즈에 맞출 수 있는 제품군을 선별하여 광고를 타켓팅.중점

  > 리뷰의 고객 정보를 대상으로 보았을때, 피부색, 머리색, 눈색, 피부 타입 등 다양한 고객 타입이 있었지만 소수의 고객이 다수.
