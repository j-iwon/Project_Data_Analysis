### üìãRFM
- Recency : transaction_date ÏµúÏã†Ïàú
- Frequency : transaction_qty
- Monetary: transaction_qty * unit_price = total_price

#### Recency

```python
print(shop_df['transaction_date'].min())
print(shop_df['transaction_date'].max())
```

```python
import pandas as pd

# 'transaction_date' Ïó¥ÏùÑ datetime ÌòïÏãùÏúºÎ°ú Î≥ÄÌôò
shop_df['transaction_date'] = pd.to_datetime(shop_df['transaction_date'])

today = pd.to_datetime('2023-06-30')
shop_df['Recency'] = (today - shop_df['transaction_date']).dt.days

display(shop_df)
```

#### Monetary

```python
shop_df.loc[:, 'total_price'] = shop_df['transaction_qty'] * shop_df['unit_price']
shop_df
```

```python
shop_df.loc[:, 'Monetary'] =  shop_df['total_price']
shop_df
```

#### Frequency = Quantity

```python
shop_df.loc[:, 'Frequency'] = shop_df['transaction_qty']
shop_df
```

#### RFM Ï†ïÍ∑úÌôî Î∞è Ï†êÏàò Ìï©ÏÇ∞

```python
from sklearn.preprocessing import MinMaxScaler

rfm_scale = MinMaxScaler()
shop_df[['Recency', 'Frequency', 'Monetary']] = rfm_scale.fit_transform(shop_df[['Recency', 'Frequency', 'Monetary']])
shop_df[['Recency', 'Frequency', 'Monetary']].mean()
```

```python
shop_df['Recency'] = 1- shop_df['Recency']
shop_df
```

```python
# Ï†ÑÏ≤¥ Ï†êÏàò Ìï©ÏÇ∞
shop_df.loc[:, 'total_score'] = shop_df['Recency'] + shop_df['Frequency'] + shop_df['Monetary']
shop_df
```

#### Îì±Í∏â ÎÇòÎàÑÍ∏∞

```python
import numpy as np

l1, l2, l3, l4 = np.percentile(shop_df['total_score'], [15, 35, 70, 90])
print(l1, l2, l3, l4)
```

```python
def get_level(x):
    if x <= l1:
        return 5
    if x <= l2:
        return 4
    if x <= l3:
        return 3
    if x <= l4:
        return 2
    return 1
```

```python
shop_df['Level'] = shop_df['total_score'].apply(get_level)
shop_df
```

```python
shop_df['Level'] = shop_df['Level'].replace([5, 4, 3, 2, 1], 
                                                ['Bronze', 'Silver', 'Gold', 'Diamond', 'Platinum'])
shop_df
```

```python
import matplotlib.pyplot as plt

def pie_chart(column_name=None, value_name=None):
    # column_nameÍ≥º value_nameÏóê Ìï¥ÎãπÌïòÎäî Îç∞Ïù¥ÌÑ∞Ïùò ÎπàÎèÑÎ•º Í≥ÑÏÇ∞Ìï©ÎãàÎã§.
    counts = shop_df.groupby(column_name)[value_name].sum()

    # ÌååÏù¥ Ï∞®Ìä∏Î•º Í∑∏Î¶ΩÎãàÎã§.
    plt.figure(figsize=(3, 3))
    plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=140)
    plt.axis('equal')  # ÏõêÏùÑ ÎèôÍ∑∏ÎûóÍ≤å Ïú†ÏßÄÌï©ÎãàÎã§.
    plt.title(f'Pie Chart for {column_name} by {value_name}')
    plt.show()

# Ìï®ÏàòÎ•º Ìò∏Ï∂úÌïòÏó¨ store_location Î≥Ñ unit_price ÌåêÎß§ÎüâÏùÑ ÌååÏù¥ Ï∞®Ìä∏Î°ú ÏãúÍ∞ÅÌôîÌï©ÎãàÎã§.
pie_chart(column_name="store_location", value_name="unit_price")
```

#### üìå ÏßÄÏ†êÎ≥Ñ Îß§Ï∂ú Î∂ÑÏÑù
- ÏßÄÏ†êÎ≥ÑÎ°ú Îß§Ï∂úÏù¥ ÎπÑÏä∑Ìï®

```python
import seaborn as sns
import matplotlib.pyplot as plt

def bar_plot(column_name=None, hue=None):
    # column_nameÍ≥º hueÏóê Ìï¥ÎãπÌïòÎäî Îç∞Ïù¥ÌÑ∞Ïùò ÎπàÎèÑÎ•º Í≥ÑÏÇ∞Ìï©ÎãàÎã§.
    counts = shop_df.groupby([column_name, hue]).size().unstack()

    # ÎßâÎåÄ Í∑∏ÎûòÌîÑÎ•º Í∑∏Î¶ΩÎãàÎã§.
    counts.plot(kind='bar', stacked=True, figsize=(5, 5))
    plt.title(f'Bar Plot for {column_name} by {hue}')
    plt.xlabel(column_name)
    plt.ylabel('Frequency')
    plt.legend(title=hue)
    plt.xticks(rotation=360)
    plt.show()

# Ìï®ÏàòÎ•º Ìò∏Ï∂úÌïòÏó¨ store_location Î≥Ñ Îì±Í∏â Î∂ÑÌè¨Î•º ÎßâÎåÄ Í∑∏ÎûòÌîÑÎ°ú ÏãúÍ∞ÅÌôîÌï©ÎãàÎã§.
bar_plot(column_name="store_location", hue="Level")
```

```python
import seaborn as sns

# ÌååÏä§ÌÖîÌÜ§ ÏÉâÏÉÅ ÌåîÎ†àÌä∏ ÏÉùÏÑ±
pastel_palette = sns.color_palette("muted")

# Ï†úÌíà Ïπ¥ÌÖåÍ≥†Î¶¨Î≥Ñ ÏÉâÏÉÅ ÏÑ§Ï†ï
palette = [pastel_palette[i % len(pastel_palette)] for i in range(9)]

# Í∞Å Îì±Í∏âÎ≥Ñ Ï†úÌíà Ïπ¥ÌÖåÍ≥†Î¶¨Ïùò Î∂ÑÌè¨ Í≥ÑÏÇ∞
category_distribution = shop_df.groupby('Level')['product_category'].value_counts(normalize=True).unstack().fillna(0)

# ÏãúÍ∞ÅÌôî
category_distribution.plot(kind='bar', stacked=True, figsize=(8, 6), color=palette)
plt.title('Preference of Product Categories by Level')
plt.xlabel('Level')
plt.ylabel('Proportion')
plt.legend(title='Product Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=360)
plt.tight_layout()
plt.show()
```

#### üìå Îì±Í∏âÎ≥Ñ Íµ¨Îß§ Ïπ¥ÌÖåÍ≥†Î¶¨ Î∂ÑÏÑù
- Ï†ÑÏ≤¥Ï†ÅÏúºÎ°ú Ïª§Ìîº, Ìã∞ Ï¢ÖÎ•òÏùò ÏùåÎ£åÎ•º Ï†úÏùº ÎßéÏù¥ Íµ¨Îß§Ìï®.
- Ï†ÑÏ≤¥ Îì±Í∏âÏù¥ ÏùåÎ£å Îã§ÏùåÏúºÎ°ú Î≤†Ïù¥Ïª§Î¶¨ ÌåêÎß§ ÎπÑÏú®Ïù¥ ÎÜíÏúºÎÇò, ÌîåÎ†àÌã∞ÎÑò Îì±Í∏âÏùÄ Î≤†Ïù¥Ïª§Î¶¨Î•º Í±∞Ïùò Íµ¨Îß§ÌïòÏßÄ ÏïäÏùå.
- ÌîåÎ†àÌã∞ÎÑò Îì±Í∏âÏùÄ ÏùåÎ£å Ïù¥Ïô∏Ïùò Îã§Î•∏ Ï†úÌíàÎì§ÏùÑ Í±∞Ïùò Íµ¨Îß§ÌïòÏßÄ ÏïäÏùå.

> Ïª§Ìîº, Ìã∞, Ï¥àÏΩî ÏùåÎ£å, Î≤†Ïù¥Ïª§Î¶¨Î•º Ï†úÏô∏ÌïòÍ≥† Ïπ¥ÌéòÏùò Flavours(ÏãúÏ¶å ÌÖåÎßàÎ≥ÑÎ°ú ÎÇòÏò§Îäî Î©îÎâ¥)Ïùò ÎπÑÏ§ëÏù¥ ÌÅ∞ Í≤ÉÏùÑ Î≥º Ïàò ÏûàÎã§.
> Ïù¥Îäî ÌÉÄ Î∏åÎûúÎìúÏôÄ Ï∞®Î≥ÑÌôîÎêú Ï†úÌíàÏùÑ Ï†ÑÎûµÏúºÎ°ú Í≥†Í∞ùÏùò Î∞©Î¨∏ÏùÑ Ïù¥ÎÅå Ïàò ÏûàÎäî ÎßàÏºÄÌåÖÏùÑ Ìï† Ïàò ÏûàÎã§.
> ÎòêÌïú ÌîåÎ†àÌã∞ÎÑò Í≥†Í∞ùÏùÑ ÎåÄÏÉÅÏúºÎ°ú ÏùåÎ£åÏôÄ Î≤†Ïù¥Ïª§Î¶¨Î•º Ìï®Íªò Íµ¨Îß§ÌñàÏùÑ Ïãú Ïø†Ìè∞Ïù¥ÎÇò Ìï†Ïù∏ÏùÑ Ï†úÍ≥µÌïúÎã§Î©¥ ÌîåÎ†àÌã∞ÎÑò Îì±Í∏âÏóêÍ≤å Îß§Ï∂ú ÏÉÅÏäπÏùÑ Í∏∞ÎåÄÌï† Ïàò ÏûàÎã§.

```python
import matplotlib.pyplot as plt
import seaborn as sns

def preferred_product_category_by_level(level):
    # ÏßÄÏ†ïÎêú Îì±Í∏âÏóê Ìï¥ÎãπÌïòÎäî Îç∞Ïù¥ÌÑ∞ ÌïÑÌÑ∞ÎßÅ
    filtered_data = shop_df[shop_df['Level'] == level]

    # Îì±Í∏âÎ≥ÑÎ°ú product_categoryÏùò ÎπàÎèÑÎ•º Í≥ÑÏÇ∞
    category_counts = filtered_data['product_type'].value_counts()

    # ÎßâÎåÄ Í∑∏ÎûòÌîÑÎ°ú ÏãúÍ∞ÅÌôî
    plt.figure(figsize=(10, 5))
    sns.barplot(x=category_counts.index, y=category_counts.values, palette="pastel")
    plt.title(f'Preferred Product Categories for {level} Customers')
    plt.xlabel('product_type')
    plt.ylabel('Frequency')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# Îì±Í∏â ÏàúÏÑú ÏÑ§Ï†ï
level_order = ['Bronze', 'Silver', 'Gold', 'Diamond', 'Platinum']

# Îì±Í∏â ÏàúÏÑúÏóê Îî∞Îùº Í∑∏ÎûòÌîÑÎ•º Í∑∏Î¶¨Îäî Ìï®Ïàò Ï†ïÏùò
def visualize_preferred_product_categories():
    for level in level_order:
        preferred_product_category_by_level(level)

# Î™®Îì† Îì±Í∏âÎ≥ÑÎ°ú Ï¢ãÏïÑÌïòÎäî product_categoryÏùò ÏàúÏúÑÎ•º ÏãúÍ∞ÅÌôîÌï©ÎãàÎã§.
visualize_preferred_product_categories()
```

#### üìå Îì±Í∏âÎ≥Ñ ÏÑ†Ìò∏ Ï†úÌíà Î∂ÑÏÑù
- Ï†Ñ Îì±Í∏â Í≥†Í∞ùÏù¥ Í∞ÄÏû• ÏÑ†Ìò∏ÌïòÎäî Ï†úÌíàÏùÄ ÏßÅÏ†ë ÎÇ¥Î¶∞ Ï∞®Ïù¥Ìã∞ÏôÄ ÎìúÎ¶Ω Ïª§Ìîº Î∞îÎ¶¨Ïä§ÌÉÄÍ∞Ä ÎßåÎì† Ïª§Ìîº.
