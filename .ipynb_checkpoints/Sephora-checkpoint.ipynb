```python
# Ìï†Ïù∏Ïú® Ïª¨Îüº ÏÉùÏÑ± (ÏÜåÏàòÏ†ê Ï≤´Ïß∏ ÏûêÎ¶¨ÍπåÏßÄ)
sp_copy_df['Discount_rate'] = round(((sp_copy_df['Original_price'] - sp_copy_df['Price']) / sp_copy_df['Original_price']) * 100, 1)

# Ìï†Ïù∏Ïú® Ïª¨ÎüºÏùÑ Í∞ÄÍ≤© ÏòÜÏúºÎ°ú ÏòÆÍ∏∞Í∏∞
# discount_rate_column = sp_df.pop('discount_rate')
# sp_df.insert(sp_df.columns.get_loc('original_price') + 1, 'discount_rate', discount_rate_column)

sp_copy_df
```

#### üìãÎç∞Ïù¥ÌÑ∞ ÏãúÍ∞ÅÌôî

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Ïπ¥ÌÖåÍ≥†Î¶¨Î≥Ñ Í≥†Í∞ù ÏÑ†Ìò∏ÎèÑ, Î¶¨Î∑∞Ïàò, ÌèâÏ†ê
 
# ÏÑ†Ìò∏ÎèÑ, Î¶¨Î∑∞Ïàò Ï†ïÍ∑úÌôî Ìï®Ïàò
def normalize_likes(df):
    df['Normalized_Like'] = df['Like'] / df['Product_counts']
    return df
    
def normalize_reviews(df):
    df['Normalized_Review'] = df['Review_counts'] / df['Product_counts']
    return df

# Ïπ¥ÌÖåÍ≥†Î¶¨Î≥ÑÎ°ú Ï†úÌíà Ïàò Í≥ÑÏÇ∞
product_counts = sp_copy_df.groupby('Majorcategory').size().reset_index(name='Product_counts')

# Ï†ïÍ∑úÌôîÎêú ÏÑ†Ìò∏ÎèÑ Í≥ÑÏÇ∞
normalized_likes_df = sp_copy_df.merge(product_counts, on='Majorcategory')
normalized_likes_df = normalize_likes(normalized_likes_df)

# Ï†ïÍ∑úÌôîÎêú Î¶¨Î∑∞Ïàò Í≥ÑÏÇ∞
normalized_reviews_df = sp_copy_df.merge(product_counts, on='Majorcategory')
normalized_reviews_df = normalize_reviews(normalized_reviews_df)

# ÌèâÏ†êÏùò ÌèâÍ∑†Í∞í Í≥ÑÏÇ∞
category_mean_grade = sp_copy_df.groupby('Majorcategory')['Grade'].mean().reset_index()

# Majorcategory ÏàúÏÑú Ï†ïÏùò
order = ['Perfume', 'FaceMakeup', 'LipMakeup', 'EyesMakeup', 
        'SkinCare', 'HairCare', 'Body', 'Tool', 
        'HealthCare', 'Other']

# Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
fig, axes = plt.subplots(3, 1, figsize=(10, 10))

# Í∏∞Î≥∏ Ìè∞Ìä∏ ÌÅ¨Í∏∞ ÏÑ§Ï†ï
plt.rcParams.update({'font.size': 8})

# Ï†ïÍ∑úÌôîÎêú ÏÑ†Ìò∏ÎèÑ ÏãúÍ∞ÅÌôî
sns.barplot(ax=axes[0], x='Majorcategory', y='Normalized_Like', data=normalized_likes_df, palette='Paired', errwidth=0.9, order=order)  
axes[0].set_xlabel('Category')
axes[0].set_ylabel('Like')

# Ï†ïÍ∑úÌôîÎêú Î¶¨Î∑∞Ïàò ÏãúÍ∞ÅÌôî
sns.barplot(ax=axes[1], x='Majorcategory', y='Normalized_Review', data=normalized_reviews_df, palette='Paired', errwidth=0.9, order=order)
axes[1].set_xlabel('Category') 
axes[1].set_ylabel('Review')

# ÌèâÏ†êÏùò ÌèâÍ∑† ÏãúÍ∞ÅÌôî
sns.barplot(ax=axes[2], x='Majorcategory', y='Grade', data=category_mean_grade, palette='Paired', errwidth=0.9, order=order)
axes[2].set_xlabel('Category')
axes[2].set_ylabel('Grade')

# Í∑∏ÎûòÌîÑ Ï∂úÎ†•
# plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Ï£ºÏñ¥ÏßÑ Îç∞Ïù¥ÌÑ∞ÏÖãÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ ÏÑ†Ìò∏ÎèÑ, Î¶¨Î∑∞Ïàò, ÌèâÏ†êÏùÑ ÌîºÏñ¥Ïä® ÏÉÅÍ¥Ä Í≥ÑÏàòÎ°ú Í≥ÑÏÇ∞
correlation_matrix = sp_copy_df[['Like', 'Review_counts', 'Grade']].corr()

# ÏÉÅÍ¥Ä ÌñâÎ†¨ ÏãúÍ∞ÅÌôî
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='vlag', fmt=".2f", linewidths=.5)
# plt.title('Correlation Matrix of Like, Review_counts, Grade')  # Í∑∏ÎûòÌîÑ Ï†úÎ™© ÏÑ§Ï†ï
plt.show()
```

```python
# 'Majorcategory'Í∞Ä 'FaceMakeup', 'LipMakeup', 'EyesMakeup'Ïù∏ ÌñâÎì§ÏùÑ ÌïÑÌÑ∞ÎßÅÌïòÏó¨ ÏÉàÎ°úÏö¥ Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ ÏÉùÏÑ±
makeup_df = sp_copy_df[sp_copy_df['Majorcategory'].isin(['FaceMakeup', 'LipMakeup', 'EyesMakeup'])]

# 'Brand' Ïó¥ÏùÑ ÏÑ†ÌÉùÌïòÏó¨ Ï∂úÎ†•
makeup_df = makeup_df['Brand'].unique()
```

```python
# makeup_dfÏóê Ìè¨Ìï®ÎêòÎäî BrandÏù¥Î©¥ÏÑú MajorcategoryÍ∞Ä FaceMakeup, LipMakeup, EyesMakeupÏù∏ Ìñâ ÏÑ†ÌÉù
mp_df = sp_copy_df[(sp_copy_df['Brand'].isin(makeup_df)) & 
                              (sp_copy_df['Majorcategory'].isin(['FaceMakeup', 'LipMakeup', 'EyesMakeup']))]

# Í≤∞Í≥º ÌôïÏù∏
# display(mp_df)
```

```python
# Like Ïó¥ÏùÑ Í∏∞Ï§ÄÏúºÎ°ú ÎÇ¥Î¶ºÏ∞®ÏàúÏúºÎ°ú Ï†ïÎ†¨ÌïòÏó¨ ÏÉÅÏúÑ 10Í∞ú Ìñâ ÏÑ†ÌÉù
top_10_likes = mp_df.sort_values(by='Like', ascending=False).head(10)

# Í≤∞Í≥º ÌôïÏù∏
# display(top_10_likes)
```

```python
# Like Ïó¥ÏùÑ Í∏∞Ï§ÄÏúºÎ°ú ÎÇ¥Î¶ºÏ∞®ÏàúÏúºÎ°ú Ï†ïÎ†¨ÌïòÏó¨ ÏÉÅÏúÑ 10Í∞ú Ìñâ ÏÑ†ÌÉù
top_10_reviews = mp_df.sort_values(by='Review_counts', ascending=False).head(10)

# Í≤∞Í≥º ÌôïÏù∏
display(top_10_reviews)
```

```python
# top_10_reviewsÏôÄ top_10_likesÏóêÏÑú Í≤πÏπòÎäî NameÏùÑ Í∞ÄÏßÄÎäî Îç∞Ïù¥ÌÑ∞ ÌïÑÌÑ∞ÎßÅ
common_names = set(top_10_reviews['Name']).intersection(set(top_10_likes['Name']))
filtered_data = sp_copy_df[sp_copy_df['Name'].isin(common_names)]

# Í≤∞Í≥º ÌôïÏù∏
display(filtered_data)
```

```python
# Í¥ÄÏã¨ÎèÑÏôÄ Î¶¨Î∑∞Ïàò ÏÉÅÏúÑÏóê ÎìúÎäî Î©îÏù¥ÌÅ¨ÏóÖ Ïπ¥ÌÖåÍ≥†Î¶¨ Ï†úÌíà
display(filtered_data[['Brand', 'Name', 'Subcategory', 'Majorcategory']])
```
