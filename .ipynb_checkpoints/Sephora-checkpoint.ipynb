#### 📋데이터 전처리
- 불필요한 컬럼 삭제하기 size, URL, MarketingFlags_content, options, details, ingredients, how_to_use  
- 컬럼명 수정하기 rating > Grade , category > Subcategory, number_of_reviews > Review_counts , love > Like , value_price > Original_price  
- Marketingflags bool 타입 int64로 변경하기  
- 할인율을 나타내는 Discount_rate 컬럼 생성하기
- 카테고리 세분화하기


```python
# 전처리 전 원본 데이터 복사하기
sp_copy_df = sp_df.copy()
sp_copy_df
```

```python
# 불필요한 컬럼 삭제하기
sp_copy_df = sp_copy_df.drop(columns=['size', 'URL', 'MarketingFlags_content', 'options', 'details', 'ingredients', 'how_to_use'])
sp_copy_df
```

```python
# 컬럼명 변경하기
sp_copy_df.columns = ['Id', 'Brand', 'Subcategory', 'Name', 'Grade', 'Review_counts', 'Like', 'Price', 'Original_price', 'Marketingflags', 'Online_only', 'Exclusive', 'Limited_edition', 'Limited_time_offer']
sp_copy_df
```

```python
# bool 값을 0 또는 1로, int64 타입으로 변환
sp_copy_df['Marketingflags'] = sp_copy_df['Marketingflags'].astype(int)
sp_copy_df
```

```python
# 할인율 컬럼 생성 (소수점 첫째 자리까지)
sp_copy_df['Discount_rate'] = round(((sp_copy_df['Original_price'] - sp_copy_df['Price']) / sp_copy_df['Original_price']) * 100, 1)

# 할인율 컬럼을 가격 옆으로 옮기기
# discount_rate_column = sp_df.pop('discount_rate')
# sp_df.insert(sp_df.columns.get_loc('original_price') + 1, 'discount_rate', discount_rate_column)

sp_copy_df
```

#### 📋데이터 시각화

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 카테고리별 고객 선호도, 리뷰수, 평점
 
# 선호도, 리뷰수 정규화 함수
def normalize_likes(df):
    df['Normalized_Like'] = df['Like'] / df['Product_counts']
    return df
    
def normalize_reviews(df):
    df['Normalized_Review'] = df['Review_counts'] / df['Product_counts']
    return df

# 카테고리별로 제품 수 계산
product_counts = sp_copy_df.groupby('Majorcategory').size().reset_index(name='Product_counts')

# 정규화된 선호도 계산
normalized_likes_df = sp_copy_df.merge(product_counts, on='Majorcategory')
normalized_likes_df = normalize_likes(normalized_likes_df)

# 정규화된 리뷰수 계산
normalized_reviews_df = sp_copy_df.merge(product_counts, on='Majorcategory')
normalized_reviews_df = normalize_reviews(normalized_reviews_df)

# 평점의 평균값 계산
category_mean_grade = sp_copy_df.groupby('Majorcategory')['Grade'].mean().reset_index()

# Majorcategory 순서 정의
order = ['Perfume', 'FaceMakeup', 'LipMakeup', 'EyesMakeup', 
        'SkinCare', 'HairCare', 'Body', 'Tool', 
        'HealthCare', 'Other']

# 그래프 그리기
fig, axes = plt.subplots(3, 1, figsize=(10, 10))

# 기본 폰트 크기 설정
plt.rcParams.update({'font.size': 8})

# 정규화된 선호도 시각화
sns.barplot(ax=axes[0], x='Majorcategory', y='Normalized_Like', data=normalized_likes_df, palette='Paired', errwidth=0.9, order=order)  
axes[0].set_xlabel('Category')
axes[0].set_ylabel('Like')

# 정규화된 리뷰수 시각화
sns.barplot(ax=axes[1], x='Majorcategory', y='Normalized_Review', data=normalized_reviews_df, palette='Paired', errwidth=0.9, order=order)
axes[1].set_xlabel('Category') 
axes[1].set_ylabel('Review')

# 평점의 평균 시각화
sns.barplot(ax=axes[2], x='Majorcategory', y='Grade', data=category_mean_grade, palette='Paired', errwidth=0.9, order=order)
axes[2].set_xlabel('Category')
axes[2].set_ylabel('Grade')

# 그래프 출력
# plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

정규화한 이유

상대적 비교 용이성: 제품 카테고리 간의 크기가 다르면 각 카테고리의 리뷰 수를 직접 비교하기 어려울 수 있습니다. 정규화를 통해 각 카테고리의 제품 수에 상관없이 상대적인 리뷰 수를 비교할 수 있습니다.

공정한 비교: 제품 수가 많은 카테고리에는 일반적으로 더 많은 리뷰가 있을 수 있습니다. 그러나 제품 수에 관계없이 각 카테고리의 리뷰 수를 정규화하면 각 카테고리 간의 리뷰 비율을 공정하게 비교할 수 있습니다.

패턴 파악: 정규화된 리뷰 수를 사용하면 특정 카테고리에서의 리뷰가 다른 카테고리에 비해 상대적으로 얼마나 많은지 더 잘 이해할 수 있습니다. 이를 통해 특정 제품 카테고리에 대한 소비자 관심도나 반응에 대한 패턴을 파악할 수 있습니다.

데이터 비교 가능성: 정규화된 데이터는 다른 데이터 집합과 비교할 때 더 유용합니다. 예를 들어, 여러 기업이나 시장 간에 비교할 때 카테고리별로 정규화된 리뷰 수를 사용하여 산업 트렌드를 비교할 수 있습니다.

이러한 분석 결과를 통해 마케팅 전략 수립이나 제품 포지셔닝에 도움을 줄 수 있습니다.  
예를 들어, 선호도가 낮은 카테고리의 제품에 대한 마케팅 노력을 강화하거나,  
높은 평점을 받는 제품을 중심으로 제품 라인업을 조정하는 등의 전략을 수립할 수

카테고리 선호도, 리뷰수, 비교 : 이러한 결과를 통해 어떤 카테고리의 제품이 가장 선호되는지, 어떤 카테고리의 제품이 가장 많은 관심을 받고 있는지 파악할 수 있습니다. 이는 제품 포지셔닝이나 마케팅 전략을 결정하는 데 도움이 될 수 있습니다.

평점이 분별력이 없다면 리뷰수를 활용하여 제품 라인업을 강조하는 전략을 고려할 수 있습니다. 리뷰수는 소비자들의 관심과 제품에 대한 인지도를 나타내는 중요한 지표 중 하나이기 때문에, 높은 리뷰수를 가진 제품은 다수의 고객들로부터 관심을 받고 있을 가능성이 높습니다.

그러나 이 전략을 채택하기 전에 몇 가지 고려해야 할 점이 있습니다. 첫째, 리뷰수가 높다고 해서 반드시 제품이 좋다는 것을 의미하지는 않습니다. 리뷰수는 제품을 구매한 소비자들의 수를 나타내는 지표이지만, 이들의 평가는 다를 수 있습니다. 둘째, 리뷰수는 제품이 출시된 시기나 마케팅 활동 등 여러 요인에 영향을 받을 수 있으므로, 단순히 리뷰수만을 보고 제품의 품질이나 성능을 판단하기는 어려울 수 있습니다.

따라서 리뷰수를 활용한 전략을 채택할 때에는 다른 지표들과 함께 고려해야 합니다. 예를 들어, 리뷰수가 높은 제품 중에서도 선호도가 높은 제품을 선별하여 강조하는 전략이나, 리뷰수가 높은 제품들 간의 평균 평점을 비교하여 상대적인 품질을 고려하는 것이 좋습니다. 종합적으로 제품의 다양한 측면을 고려하여 전략을 수립하는 것이 중요합니다. 있습니다.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 주어진 데이터셋을 사용하여 선호도, 리뷰수, 평점을 피어슨 상관 계수로 계산
correlation_matrix = sp_copy_df[['Like', 'Review_counts', 'Grade']].corr()

# 상관 행렬 시각화
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='vlag', fmt=".2f", linewidths=.5)
# plt.title('Correlation Matrix of Like, Review_counts, Grade')  # 그래프 제목 설정
plt.show()
```

상관관계 분석

```python
# 'Majorcategory'가 'FaceMakeup', 'LipMakeup', 'EyesMakeup'인 행들을 필터링하여 새로운 데이터프레임 생성
makeup_df = sp_copy_df[sp_copy_df['Majorcategory'].isin(['FaceMakeup', 'LipMakeup', 'EyesMakeup'])]

# 'Brand' 열을 선택하여 출력
makeup_df = makeup_df['Brand'].unique()
```

```python
# makeup_df에 포함되는 Brand이면서 Majorcategory가 FaceMakeup, LipMakeup, EyesMakeup인 행 선택
mp_df = sp_copy_df[(sp_copy_df['Brand'].isin(makeup_df)) & 
                              (sp_copy_df['Majorcategory'].isin(['FaceMakeup', 'LipMakeup', 'EyesMakeup']))]

# 결과 확인
# display(mp_df)
```

```python
# Like 열을 기준으로 내림차순으로 정렬하여 상위 10개 행 선택
top_10_likes = mp_df.sort_values(by='Like', ascending=False).head(10)

# 결과 확인
# display(top_10_likes)
```

```python
# Like 열을 기준으로 내림차순으로 정렬하여 상위 10개 행 선택
top_10_reviews = mp_df.sort_values(by='Review_counts', ascending=False).head(10)

# 결과 확인
display(top_10_reviews)
```

```python
# top_10_reviews와 top_10_likes에서 겹치는 Name을 가지는 데이터 필터링
common_names = set(top_10_reviews['Name']).intersection(set(top_10_likes['Name']))
filtered_data = sp_copy_df[sp_copy_df['Name'].isin(common_names)]

# 결과 확인
display(filtered_data)
```

```python
# 관심도와 리뷰수 상위에 드는 메이크업 카테고리 제품
display(filtered_data[['Brand', 'Name', 'Subcategory', 'Majorcategory']])
```

할인율: 가격과 원래 가격 간의 할인율(Discount_rate)을 계산하여 할인 정책이 고객들의 관심을 어떻게 유발하는지를 살펴봅니다. 높은 할인율이 고객들의 관심을 끄는지, 실제로 구매로 이어지는지 등을 분석할 수 있습니다.

고객들의 관심도: 고객들의 관심도(Like)를 가격과 할인율과 비교하여, 제품의 가격이나 할인 정책이 관심도에 미치는 영향을 조사합니다. 예를 들어, 가격이 낮은 제품일수록 관심도가 높은지, 할인율이 높은 제품일수록 관심도가 높은지 등을 살펴볼 수 있습니다.

마케팅 플래그: 마케팅 플래그(Marketingflags)나 다른 마케팅 관련 변수들을 활용하여 특정 마케팅 전략이 제품의 가격, 할인율, 관심도에 미치는 영향을 분석할 수 있습니다. 예를 들어, 온라인 전용 제품인 경우 할인율이 더 높은지, 한정판이나 한정 기간 세일 제품의 관심도가 더 높은지 등을 조사할 수 있습니다.

제품 카테고리 비교: 각 카테고리별로 가격, 할인율, 관심도의 차이를 분석하여, 제품의 특성에 따른 고객들의 반응을 파악할 수 있습니다. 예를 들어, 스킨케어 제품의 경우 가격이 높더라도 고객들의 관심도가 높은지, 메이크업 제품의 경우 할인율이 높을수록 관심도가 높은지 등을 비교합니다.
