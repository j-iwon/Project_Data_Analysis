### üéÄRFM
- Recency : invoice_date ÏµúÏã†Ïàú
- Frequency : quantity
- Monetary: quantity * price = total_price

#### üìçRecency


```python
# ÎÇ†Ïßú Îç∞Ïù¥ÌÑ∞ Î≥ÄÌôò
cs_df['invoice_date']= pd.to_datetime(cs_df['invoice_date'], dayfirst=True)
```

```python
print(cs_df['invoice_date'].min())
print(cs_df['invoice_date'].max())
```

```python
today = pd.to_datetime('2023.03.10')
cs_df.loc[:, 'Recency'] = cs_df['invoice_date'].apply(lambda x: (today-x).days)
cs_df
```

#### üìçMonetary

```python
cs_df.loc[:, 'total_price'] = cs_df['quantity'] * cs_df['price']
cs_df
```

```python
cs_df.loc[:, 'Monetary'] =  cs_df['total_price']
cs_df
```

#### üìçFrequency = Quantity

```python
cs_df.loc[:, 'Frequency'] = cs_df['quantity']
cs_df
```

#### üìçRFM Ï†ïÍ∑úÌôî Î∞è Ï†êÏàò Ìï©ÏÇ∞

```python
from sklearn.preprocessing import MinMaxScaler

rfm_scale = MinMaxScaler()
cs_df[['Recency', 'Frequency', 'Monetary']] = rfm_scale.fit_transform(cs_df[['Recency', 'Frequency', 'Monetary']])
cs_df[['Recency', 'Frequency', 'Monetary']].mean()
```

```python
# Í∞ÄÏ§ëÏπò Ï£ºÍ∏∞
cs_df['Monetary'] = cs_df['Monetary']*3
```

```python
cs_df['Recency'] = 1- cs_df['Recency']
cs_df
```

```python
# Ï†ÑÏ≤¥ Ï†êÏàò Ìï©ÏÇ∞
cs_df.loc[:, 'total_score'] = cs_df['Recency'] + cs_df['Frequency'] + cs_df['Monetary']
cs_df
```

#### Îì±Í∏âÎÇòÎàÑÍ∏∞

```python
import numpy as np

l1, l2, l3, l4 = np.percentile(cs_df['total_score'], [20, 45, 75, 90])
print(l1, l2, l3, l4)
```

```python
def get_level(x):
    if x <= l1:
        return 5
    if x <= l2:
        return 4
    if x <= l3:
        return 3
    if x <= l4:
        return 2
    return 1
```

```python
cs_df['Level'] = cs_df['total_score'].apply(get_level)
cs_df
```

```python
cs_df['Level'] = cs_df['Level'].replace([5, 4, 3, 2, 1], 
                                                ['Bronze', 'Silver', 'Gold', 'Diamond', 'Platinum'])
```

```python
cs_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

order = ['Bronze', 'Silver', 'Gold', 'Diamond', 'Platinum']

sns.countplot(x = 'Level', data=cs_df, palette='muted', order=order)
plt.title('Level Count')
plt.show()
```

  > Í≥®Îìú Îì±Í∏âÏùò Ïù∏ÏõêÏàòÎäî Ï¶ùÍ∞ÄÌïòÎäî Îç∞Ïóê ÎåÄÎπÑÌïòÏó¨ Íµ¨Îß§ Í∏àÏï°Ïù¥ ÌäπÏ†ï Î∂ÑÍ∏∞Ïóê Í∞êÏÜåÌïòÎäî Ìå®ÌÑ¥Ïù¥ ÎÇòÌÉÄÎÇòÍ≥† ÏûàÎäî Í≤ÉÏùÑ Í≥†Î†§ÌïòÏó¨, Ìï¥Îãπ Î∂ÑÍ∏∞Ïóê Í≥®Îìú Îì±Í∏â Í≥†Í∞ùÎì§ÏùÑ ÎåÄÏÉÅÏúºÎ°ú Ìïú ÌîÑÎ°úÎ™®ÏÖò Î∞è Ïù¥Î≤§Ìä∏Î•º Í∞úÏµúÌïòÏó¨ Íµ¨Îß§Î•º Ï¥âÏßÑÌï† ÌïÑÏöîÍ∞Ä ÏûàÎã§.


```python
# femaleÍ≥º maleÏóê ÎåÄÌïú Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ ÏÉùÏÑ±
female_distribution = cs_df[cs_df['gender'] == 'Female'].groupby('Level').size()
male_distribution = cs_df[cs_df['gender'] == 'Male'].groupby('Level').size()

# femaleÍ≥º male Îç∞Ïù¥ÌÑ∞Î•º ÌïòÎÇòÏùò Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏúºÎ°ú Ìï©ÏπòÍ∏∞
gender_distribution = pd.concat([female_distribution, male_distribution], axis=1, keys=['Female', 'Male']).fillna(0)

# Í∞Å Ïó∞Î†π Î≥Ñ ÏÑ±Î≥Ñ ÎπÑÏú® Í≥ÑÏÇ∞
gender_ratio_by_age = cs_df.groupby(['age', 'gender']).size().unstack().reset_index()
gender_ratio_by_age['total'] = gender_ratio_by_age['Female'] + gender_ratio_by_age['Male']
gender_ratio_by_age['female_ratio'] = gender_ratio_by_age['Female'] / gender_ratio_by_age['total']
gender_ratio_by_age['male_ratio'] = gender_ratio_by_age['Male'] / gender_ratio_by_age['total']

# Í∞Å ÏáºÌïëÎ™∞ Î≥Ñ ÏÑ±Î≥Ñ ÎπÑÏú® Í≥ÑÏÇ∞
gender_ratio_by_store = cs_df.groupby(['shopping_mall', 'gender']).size().unstack().reset_index()
gender_ratio_by_store['total'] = gender_ratio_by_store['Female'] + gender_ratio_by_store['Male']
gender_ratio_by_store['female_ratio'] = gender_ratio_by_store['Female'] / gender_ratio_by_store['total']
gender_ratio_by_store['male_ratio'] = gender_ratio_by_store['Male'] / gender_ratio_by_store['total']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import seaborn as sns

# Ï∂ï Í∏ÄÏûê Ìè∞Ìä∏ ÏÇ¨Ïù¥Ï¶à
plt.rc('axes', labelsize=20)

# Í∑∏ÎûòÌîÑ ÌÅ¨Í∏∞ ÏÑ§Ï†ï
plt.figure(figsize=(20, 6))

# Í∑∏ÎûòÌîÑ Ï†úÎ™© Î∞è Îç∞Ïù¥ÌÑ∞ ÏÑ§Ï†ï
graphs = [
    {'title': 'Gender Distribution by Level', 'data': gender_distribution},
    {'title': 'Gender Ratio by Shopping Mall', 'data': gender_ratio_by_store},
    {'title': 'Gender Ratio by Age', 'data': gender_ratio_by_age}
]

# Î∞òÎ≥µÎ¨∏ÏùÑ ÌÜµÌï¥ Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
for i, graph in enumerate(graphs, 1):
    plt.subplot(1, 3, i)
    if graph['title'] == 'Gender Distribution by Level':
        graph['data'].plot(kind='bar', ax=plt.gca(), color=['pink', 'blue'], stacked=True)
        plt.xticks(rotation=45)
    elif graph['title'] == 'Gender Ratio by Shopping Mall':
        sns.barplot(data=graph['data'], x='shopping_mall', y='female_ratio', color='pink', label='Female')
        sns.barplot(data=graph['data'], x='shopping_mall', y='male_ratio', color='blue', bottom=graph['data']['female_ratio'], label='Male')
    else:
        sns.barplot(data=graph['data'], x='age', y='female_ratio', color='pink', label='Female')
        sns.barplot(data=graph['data'], x='age', y='male_ratio', color='blue', bottom=graph['data']['female_ratio'], label='Male')
    plt.xlabel('')  # xÏ∂ï ÎùºÎ≤® ÎπÑÏö∞Í∏∞
    plt.ylabel('Gender Ratio')
    plt.title(graph['title'])
    plt.legend()

# Î†àÏù¥ÏïÑÏõÉ Ï°∞Ï†ï
plt.tight_layout()

# Í∑∏ÎûòÌîÑ ÌëúÏãú
plt.show()
```

```python
# Ïó∞Î†πÎåÄ Íµ¨Î∂Ñ Ìï®Ïàò
def age_group(age):
    if age < 20:
        return 'Under 20'
    elif age < 30:
        return '20s'
    elif age < 40:
        return '30s'
    elif age < 50:
        return '40s'
    elif age < 60:
        return '50s'
    else:
        return 'Over 60'
```

```python
# Îì±Í∏âÎ≥Ñ Ïó∞Î†π Î∂ÑÌè¨ Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ ÏÉùÏÑ±
sm_distribution = cs_df.groupby('shopping_mall')['age'].apply(list).reset_index()

# Í∞Å ÏáºÌïëÎ™∞Î≥Ñ Îì±Í∏âÎ≥Ñ Ïó∞Î†πÎåÄ ÎπÑÏú® Í≥ÑÏÇ∞
sm_distribution['Age_Group'] = sm_distribution['age'].apply(lambda x: [age_group(age) for age in x])
age_counts = sm_distribution.explode('Age_Group').groupby(['shopping_mall', 'Age_Group']).size().unstack(fill_value=0)

# ÌååÏù¥ Ï∞®Ìä∏ Í∑∏Î¶¨Í∏∞
plt.figure(figsize=(20, 10))
for mall, counts in age_counts.iterrows():
    plt.subplot(2, 5, list(age_counts.index).index(mall) + 1)
    plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=90)
    plt.title(mall)

plt.tight_layout()
plt.show()
```

```python
# Îì±Í∏âÎ≥Ñ Ïó∞Î†π Î∂ÑÌè¨ Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ ÏÉùÏÑ±
item_distribution = cs_df.groupby('category')['age'].apply(list).reset_index()


# Í∞Å ÏáºÌïëÎ™∞Î≥Ñ Îì±Í∏âÎ≥Ñ Ïó∞Î†πÎåÄ ÎπÑÏú® Í≥ÑÏÇ∞
item_distribution['Age_Group'] = sm_distribution['age'].apply(lambda x: [age_group(age) for age in x])
age_counts = item_distribution.explode('Age_Group').groupby(['category', 'Age_Group']).size().unstack(fill_value=0)

# ÌååÏù¥ Ï∞®Ìä∏ Í∑∏Î¶¨Í∏∞
plt.figure(figsize=(20, 10))
for mall, counts in age_counts.iterrows():
    plt.subplot(2, 5, list(age_counts.index).index(mall) + 1)
    plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=90)
    plt.title(mall)

plt.tight_layout()
plt.show()
```

#### Í∞Å ÏáºÌïëÎ™∞Ïùò Ïπ¥ÌÖåÍ≥†Î¶¨ Íµ¨ÏÑ± ÎπÑÏú®

```python
# Í∞Å ÏáºÌïëÎ™∞ÏóêÏÑú ÌåêÎß§ÌïòÎäî Ïπ¥ÌÖåÍ≥†Î¶¨ ÌôïÏù∏
categories_by_shopping_mall = cs_df.groupby('shopping_mall')['category'].value_counts()

# ÏáºÌïëÎ™∞ Í∞úÏàò
num_shopping_malls = len(categories_by_shopping_mall.index.levels[0])

# Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(20, 10))

for i, mall in enumerate(categories_by_shopping_mall.index.levels[0]):
    row = i // 5  # Ìñâ Ïù∏Îç±Ïä§ Í≥ÑÏÇ∞
    col = i % 5   # Ïó¥ Ïù∏Îç±Ïä§ Í≥ÑÏÇ∞
    
    categories = categories_by_shopping_mall[mall]
    axes[row, col].pie(categories.values, labels=categories.index, autopct='%1.1f%%', startangle=90)
    axes[row, col].set_title(f'Categories in {mall}')

plt.tight_layout()
plt.show()
```

#### üìåÌäπÏù¥Ï†ê

- Ï†Ñ ÏáºÌïëÎ™∞Ïùò ÌåêÎß§ÌïòÎäî Ïπ¥ÌÖåÍ≥†Î¶¨Ïùò ÎπÑÏú®Ïù¥ ÎπÑÏä∑Ìï®

#### Í∞Å ÏáºÌïëÎ™∞Ïùò Î∂ÑÍ∏∞Î≥Ñ Îß§Ï∂úÏï°

```python
# Ï∂ï Í∏ÄÏûê Ìè∞Ìä∏ ÏÇ¨Ïù¥Ï¶à
plt.rc('axes', labelsize=20)

# Í∞Å Ïó∞ÎèÑÎ°ú Î∂ÑÎ¶¨
purchase_amount_2021 = cs_df[cs_df['Year'] == 2021].groupby(['Quarter', 'shopping_mall'])['total_price'].sum().reset_index()
purchase_amount_2022 = cs_df[cs_df['Year'] == 2022].groupby(['Quarter', 'shopping_mall'])['total_price'].sum().reset_index()
purchase_amount_2023 = cs_df[cs_df['Year'] == 2023].groupby(['Quarter', 'shopping_mall'])['total_price'].sum().reset_index()

# Í∞Å Ïó∞ÎèÑÎ≥Ñ ÏµúÎåÄ Îß§Ï∂úÏï° Í≥ÑÏÇ∞
max_purchase_amount_2021 = purchase_amount_2021['total_price'].max()
max_purchase_amount_2022 = purchase_amount_2022['total_price'].max()
max_purchase_amount_2023 = purchase_amount_2023['total_price'].max()

fig, axs = plt.subplots(1, 3, figsize=(20, 6))  # 1Ìñâ 3Ïó¥Ïùò ÏÑúÎ∏åÌîåÎ°Ø ÏÉùÏÑ±

# Ïó∞ÎèÑÎ≥Ñ Îß§Ï∂ú Îç∞Ïù¥ÌÑ∞ Î¶¨Ïä§Ìä∏
purchase_data_list = [purchase_amount_2021, purchase_amount_2022, purchase_amount_2023]

# Í∞Å Ïó∞ÎèÑÏóê ÎåÄÌïú Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
for i, purchase_data in enumerate(purchase_data_list):
    sns.barplot(data=purchase_data, x='Quarter', y='total_price', hue='shopping_mall', ax=axs[i])
    axs[i].set_title(f'Total Price by Quarter and Shopping Mall in {2019 + i}')
    axs[i].set_xlabel('Quarter')
    axs[i].set_ylabel('Purchase Amount')
    axs[i].set_ylim(0, max(max_purchase_amount_2021, max_purchase_amount_2022, max_purchase_amount_2023))  # ÏµúÎåÄ Îß§Ï∂úÏï°ÏúºÎ°ú y Ï∂ï Î≤îÏúÑ ÏÑ§Ï†ï
    axs[i].legend(title='Shopping Mall', bbox_to_anchor=(1, 1))

plt.tight_layout()
plt.show()
```

### Îì±Í∏âÎ≥Ñ ÏÑ†Ìò∏ ÏáºÌïëÎ™∞Í≥º ÏÜåÎπÑÎüâ

```python
# Îì±Í∏â Î™©Î°ù
levels = ['Platinum', 'Diamond', 'Gold', 'Silver', 'Bronze']

# Í∞Å Îì±Í∏âÎ≥Ñ Í∞ÄÏû• ÎßéÏù¥ Î∞©Î¨∏Ìïú ÏáºÌïëÎ™∞ Ï†ÄÏû•Ìï† ÎîïÏÖîÎÑàÎ¶¨
most_visited_malls = {}

# Î∞òÎ≥µÎ¨∏ÏùÑ ÌÜµÌï¥ Í∞Å Îì±Í∏âÎ≥ÑÎ°ú Í∞ÄÏû• ÎßéÏù¥ Î∞©Î¨∏Ìïú ÏáºÌïëÎ™∞ Ï∞æÍ∏∞
for level in levels:
    # Ìï¥Îãπ Îì±Í∏âÏùò Í≥†Í∞ù ÌïÑÌÑ∞ÎßÅ
    customers = cs_df[cs_df['Level'] == level]
    
    # Ìï¥Îãπ Îì±Í∏â Í≥†Í∞ùÏù¥ Î∞©Î¨∏Ìïú ÏáºÌïëÎ™∞Î≥Ñ Î∞©Î¨∏ ÌöüÏàò Í≥ÑÏÇ∞
    visits_by_shopping_mall = customers['shopping_mall'].value_counts()
    
    # Í∞ÄÏû• ÎßéÏù¥ Î∞©Î¨∏ÌïòÎäî ÏáºÌïëÎ™∞ Ï∞æÍ∏∞
    most_visited_mall = visits_by_shopping_mall.idxmax()
    
    # ÎîïÏÖîÎÑàÎ¶¨Ïóê Ï†ÄÏû•
    most_visited_malls[level] = most_visited_mall

# Í≤∞Í≥º Ï∂úÎ†•
display(most_visited_malls)
```

```python
# Îì±Í∏â(Level)Ïùò ÏàúÏÑú Ï†ïÏùò
level_order = ['Bronze', 'Silver', 'Gold', 'Diamond', 'Platinum']

# Í∞Å ÏáºÌïëÎ™∞ Î≥Ñ Îì±Í∏âÎ≥Ñ Íµ¨Îß§Í∏àÏï°ÏùÑ ÌîºÎ≤ó ÌÖåÏù¥Î∏îÎ°ú Ï†ïÎ¶¨ÌïòÍ≥† ÏàúÏÑú ÏßÄÏ†ï
mall_level_revenue_pivot = cs_df.pivot_table(index='shopping_mall', columns='Level', values='total_price', aggfunc='sum')
mall_level_revenue_pivot = mall_level_revenue_pivot.reindex(columns=level_order)

# ÏãúÍ∞ÅÌôî
plt.figure(figsize=(12, 8))
sns.heatmap(mall_level_revenue_pivot, cmap='YlGnBu', annot=True, fmt=".1f", linewidths=.5)
plt.title('Total Revenue by Level and Shopping Mall')
plt.xlabel('Level')
plt.ylabel('Shopping Mall')
plt.xticks(rotation=45)
plt.show()
```
